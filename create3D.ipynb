Typical workflow:
1. Generate a grid of (x, y) points with meshgrid
2. Compute z = f(x, y) at every point
3. Triangulate the surface and write the triangles to an STL

```python
import numpy as np
from stl import mesh
from skimage import measure
import matplotlib.pyplot as plt
```

```python
x = y = np.linspace(-1, 1, 100) # 100 evenly spaced numbers from -1 to 1
z = np.linspace(0, 0.5, 100) # 100 evenly spaced numbers from 0 to 0.5
x2d, y2d = np.meshgrid(x, y, indexing='xy') # turn 1D arrays into a grid for calculations
x3d, y3d, z3d = np.meshgrid(x, y, z, indexing='xy')
```
